     1. IRIS DATASET (Classification – SVM)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, log_loss
```

```python
iris = load_iris()

df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target

df.head()
```

```python
print("Shape:", df.shape)
print(df.isnull().sum())
print(df.describe())
```

```python
df_filtered = df[df['species'].isin([0,1])]

print(df_filtered['species'].value_counts())
```

```python
X = df_filtered.drop('species', axis=1)
y = df_filtered['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
model = SVC(probability=True)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)

y_prob = model.predict_proba(X_test)
loss = log_loss(y_test, y_prob)

print("Accuracy:", accuracy)
print("Log Loss:", loss)
```

      2. BOSTON HOUSING DATASET (Regression – Linear Regression)

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

boston = pd.read_csv("housing.csv")
boston.head()
```

```python
boston = pd.read_csv("housing.csv")

boston.head()
```

```python
print("Missing Values:\n", boston.isnull().sum())

# fill missing values with mean of numeric columns only
boston.fillna(boston.mean(numeric_only=True), inplace=True)
```

```python
print(boston.columns)
```

```python
correlation = boston.corr(numeric_only=True)

print(correlation['Price'].sort_values(ascending=False))   
```

```python
print(boston.columns)
```

```python
avg_rooms = boston['Avg. Area Number of Rooms'].mean()
filtered_boston = boston[boston['Avg. Area Number of Rooms'] > avg_rooms]
```

```python
x = filtered_boston.drop('Price', axis=1)
y = filtered_boston['Price']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python

X = filtered_boston.drop(['Price', 'Address'], axis=1) 
y = filtered_boston['Price']                            

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R2 Score:", r2)
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test, y_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted House Prices")
plt.show()
```

    SVM (Support Vector Machine) is used for classification because it works well for smaller datasets and finds the optimal hyperplane that separates classes with maximum margin. It performs effectively in binary classification like the filtered Iris dataset.

    Linear Regression is used for predicting continuous values such as house prices. It establishes a linear relationship between independent variables and the target variable. Model performance is evaluated using MAE, MSE, and R² score to measure prediction accuracy.
